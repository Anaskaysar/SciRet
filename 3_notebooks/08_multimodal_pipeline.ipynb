# 08 Multimodal Pipeline

Purpose: route query types and fuse text + figure evidence into answer context.

Inputs: text retrieval outputs + figure embeddings.
Output: multimodal retrieval traces under `4_results/multimodal/`.

```python
from pathlib import Path
import json

ROOT = Path('..').resolve()
out_dir = ROOT / '4_results' / 'multimodal'
out_dir.mkdir(parents=True, exist_ok=True)

trace = {
    'query': 'example scientific query',
    'router_decision': 'text_only',
    'text_hits': [],
    'figure_hits': [],
    'final_context_ids': []
}
(out_dir / 'sample_trace.json').write_text(json.dumps(trace, indent=2))
print(f'Initialized multimodal trace file in {out_dir}')
```
