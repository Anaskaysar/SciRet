# 06 Figure Extraction

Purpose: extract figure/table assets and metadata from parsed PDFs.

Input: CORD-19 parse folders/PDF assets.
Output: figure manifest under `1_data/processed/figures_manifest.parquet`.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
out_path = ROOT / '1_data' / 'processed' / 'figures_manifest.parquet'
out_path.parent.mkdir(parents=True, exist_ok=True)

df = pd.DataFrame(columns=['figure_id', 'cord_uid', 'page', 'asset_path', 'caption'])
df.to_parquet(out_path, index=False)
print(f'Initialized figure manifest -> {out_path}')
print('TODO: integrate PyMuPDF/pdfplumber extraction logic')
```
