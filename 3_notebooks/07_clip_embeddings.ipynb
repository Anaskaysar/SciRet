# 07 CLIP Embeddings

Purpose: generate figure embeddings for multimodal retrieval.

Input: `1_data/processed/figures_manifest.parquet`
Output: `1_data/embeddings/figure_embeddings.parquet`

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
fig_path = ROOT / '1_data' / 'processed' / 'figures_manifest.parquet'
out_path = ROOT / '1_data' / 'embeddings' / 'figure_embeddings.parquet'
out_path.parent.mkdir(parents=True, exist_ok=True)

assert fig_path.exists(), 'Run notebook 06 first.'
fig_df = pd.read_parquet(fig_path)

emb_df = fig_df[['figure_id']].copy()
emb_df['dim'] = 512
emb_df['vector_stub'] = 'replace_with_real_clip_vector'
emb_df.to_parquet(out_path, index=False)
print(f'Wrote {len(emb_df):,} figure embedding rows -> {out_path}')
```
